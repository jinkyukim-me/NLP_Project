```python
from random import *
```

```python
data = []

for cnt in range(365):
    i = randint(1,5)
    data.append(i)
print(data)
```

```python
a = (data.count(1) / 365) * 100
b = (data.count(2) / 365) * 100
c = (data.count(3) / 365) * 100
d = (data.count(4) / 365) * 100
e = (data.count(5) / 365) * 100
print("우울:{:0.2f}% 약간우울{:0.2f}% 평온{:0.2f}% 스마일{:0.2f}% 행복{:0.2f}%".format(a,b,c,d,e))
emotion_lst = []
emotion_lst.append(a)
emotion_lst.append(b)
emotion_lst.append(c)
emotion_lst.append(d)
emotion_lst.append(e)
```

```python
print("내일 당신의 감정 예측은 {}".format(max(emotion_lst)))
```
